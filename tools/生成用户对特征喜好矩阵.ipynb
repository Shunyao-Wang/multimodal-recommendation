```python
import numpy as np
import pandas as pd

# DATAPATH = "../data/"
DATAPATH = "F:/data/"
_sigma = 2.5 # 判断是否是积极评分的阈值
```

```python
file = open(DATAPATH + 'movieID.txt', 'r')
movieID = [int(x) for x in file]
file.close()
print(movieID)

file = open(DATAPATH + 'userID.txt', 'r')
userID = [int(x) for x in file]
file.close()
print(userID)
```

```python
df_rating = pd.read_csv(DATAPATH + "rating.csv")
print(df_rating.shape)
df_rating.head(5)
```

```python
feature = np.load(DATAPATH + "features/low_feature.npy")
print(feature.shape)
```

```python
movieOrder = [i for i in range(len(movieID))]
movie_dic = dict(zip(movieID, movieOrder))
print(movie_dic)
```

```python
# userID = [267,533]
FI_matrix = np.zeros((len(userID), feature.shape[1]))
print(FI_matrix.shape)
TR = 0.0
i = 0
for uid in userID:
    df_loc = df_rating[df_rating.userID == uid]
    TR = df_loc["rating"].sum()
    movie_list = df_loc["movieID"].tolist()
    rating_list = df_loc["rating"].tolist()
    FR = np.zeros(feature.shape[1])
    for mid, r in zip(movie_list, rating_list):
        if r < _sigma: # 判断是否为积极评分
            continue
        morder = movie_dic.get(mid)
        FR += feature[morder] * r
        FI = FR / TR
        FI_matrix[i] += FI
    i += 1
print(FI_matrix)
        
```

```python
np.save(DATAPATH + "FI_matrix.npy", FI_matrix)
```
